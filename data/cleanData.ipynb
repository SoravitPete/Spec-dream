```python
import csv

output_file_path = '/Users/felixia/Desktop/Spec-dream/data/formatted_data.py'

def format_casing(data):
    formatted_casings = []
    for i, casing in enumerate(data):
        formatted_casings.append({
            "name": casing['Name'],
            "formfactor": casing['formfactor'],
            "rgb": casing['rgb'].lower() == 'true',
            "price": int(casing['price'])
        })
    return formatted_casings

def read_csv(file_path):
    data = []
    with open(file_path, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            data.append(row)
    return data

def format_cpu(data):
    formatted_cpus = []
    for i, cpu in enumerate(data):
        formatted_cpu = {
            "name": cpu['Name'],
            "base_clock": float(cpu['base_clock'].split()[0]),
            "cores": int(cpu['cores']),
            "multi_threaded": cpu['multi_threaded'].lower() == 'true',
            "socket": cpu['socket'],
            "price": int(cpu['Price']),
            "brand": cpu['brand']
        }
        formatted_cpus.append(formatted_cpu)
    return formatted_cpus

def format_ram(data):
    formatted_ram = []
    for i, ram in enumerate(data):
        formatted_ram.append({
            "name": ram['Name'],
            "capacity": int(ram['capacity']),
            "rgb": ram['rgb'].lower() == 'true',
            "price": int(ram['Price'])
        })
    return formatted_ram

def format_gpu(data):
    formatted_gpus = []
    for i, gpu in enumerate(data):
        formatted_gpus.append({
            "name": gpu['Name'],
            # "core_clock": float(gpu['base_clock']),
            "vram": int(gpu['vram']),
            "price": int(gpu['price']),
            "brand": gpu['brand'],
            "wattage": gpu['wattage'],
            "chipset": gpu['chipset']
        })

    return formatted_gpus

def format_mainboard(data):
    formatted_mainboards = []
    for i, mb in enumerate(data):
        max_memory = int(mb['max_memory']) if mb['max_memory'] else 0
        formatted_mainboards.append({
            "name": mb['Name'],
            "formfactor": mb['formfactor'],
            "socket": mb['socket'],
            "max_memory": max_memory,
            "price": int(mb['price']),
            "memory_slot": int(mb['memory_slot']),
            "m2_slot": mb['m2_slot'],
            "brand": mb['brand']
        })
    return formatted_mainboards

def format_psu(data):
    formatted_psus = []
    for i, psu in enumerate(data):
        formatted_psus.append({
            "name": psu['Name'],
            "wattage": int(psu['wattage']),
            "price": int(psu['price'])
        })
    return formatted_psus


cpu_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/CPU-Table.csv')
ram_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/RAM-Table.csv')
gpu_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/GPU-Table.csv')
mainboard_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/MAINBOARD-Table.csv')
psu_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/PSU-Table.csv')
casing_data = read_csv('/Users/felixia/Desktop/Spec-dream/data/CASING-Table.csv')

formatted_data = {
    "CPU": format_cpu(cpu_data),
    "RAM": format_ram(ram_data),
    "GPU": format_gpu(gpu_data),
    "Motherboard": format_mainboard(mainboard_data),
    "PSU": format_psu(psu_data),
    "Casing": format_casing(casing_data),
}

print(formatted_data)

output_file_path = '/Users/felixia/Desktop/Spec-dream/data/formatted_data.txt'

def export_dict_to_txt(data_dict, file_path):
    with open(file_path, 'w') as file:
        file.write("{\n")
        for category, items in data_dict.items():
            file.write(f"  {category}: [\n")
            for item in items:
                item_str = str(item).replace("True", "true").replace("False", "false")
                file.write(f"    {item_str},\n")
            file.write("  ],\n")
        file.write("};\n")
export_dict_to_txt(formatted_data, output_file_path)
```

```python
def update_boolean_values(file_path, new_file_path):
    with open(file_path, 'r') as file:
        data = file.read()

    # Replace "true" with "True"
    data = data.replace("true", "True")
    # Replace "false" with "False"
    data = data.replace("false", "False")

    data = data.replace("'", '"')
    # Add opening and closing braces to make it a valid dictionary

    with open(new_file_path, 'w') as file:
        file.write(data)

# Example usage:
input_file_path = '/Users/felixia/Desktop/Spec-dream/data/formatted_data.txt'
output_file_path_python = '/Users/felixia/Desktop/Spec-dream/data/formatted_data_python.txt'
update_boolean_values(input_file_path, output_file_path_python)
```
